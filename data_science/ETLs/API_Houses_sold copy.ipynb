```python
#importar las librerias
import pandas as pd
import requests
import json
```

```python
#sacar un listado de los estados de EEUU
url2='https://es.wikipedia.org/wiki/Anexo:Ciudades_de_Estados_Unidos_por_poblaci%C3%B3n'
page=pd.read_html(url2)
State=page[1]
```

```python
#limpiar la url
State['Ranking.1']=(State['Ranking.1'].astype(str).str.split('-')).str[0]
```

```python
#darle formato de lista
State=State.groupby('Ranking.1')['Ciudad'].apply(list).reset_index()
State.head()
```

```python
#funcion para extraer los datos del API
def get_sold(State_code,city):
    url = "https://us-real-estate.p.rapidapi.com/sold-homes"

    querystring = {"state_code":f"{State_code}","city":f"{city}","limit":"42","offset":"0","sort":"sold_date"}

    headers = {
        "X-RapidAPI-Key": "",
	    "X-RapidAPI-Host": "us-real-estate.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Convierte la respuesta JSON en un diccionario de Python
        data = response.json()

        results = data["data"]["results"]
        results = pd.DataFrame(results)
        # Define el nombre del archivo donde se guardará la respuesta JSON
        
    else:
        results=f"the request was failed. Error code: {response.status_code}"

    return results

```

```python
#Probar el API
get_sold('AK','Anchorage')
```

```python
#hacer una lista de los estados para usarlo en la funcion de extraccion del Api

dfs=[]
for index,row in State.iterrows():
    
    State_code=row['Ranking.1']

    for city in row['Ciudad']:

        df=get_sold(State_code,city)
        

        dfs.append(df)


        
```

```python
#Añadir cada uno de los datos de cada estado al dataframe final
df_final=pd.concat(dfs,ignore_index=True)
```

```python
df_final.to_csv('Dataset_houses_sold.csv',index=False)
```
