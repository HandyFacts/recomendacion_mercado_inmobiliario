```python
import pandas as pd
import requests
import json
```

```python

def get_for_sale(State_code,city):
    url = "https://us-real-estate.p.rapidapi.com/v3/for-sale"

    querystring = {"state_code":f"{State_code}","city":f"{city}","sort":"newest","offset":"0","limit":"42"}

    headers = {
        "X-RapidAPI-Key": "xx",
        "X-RapidAPI-Host": "us-real-estate.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Convierte la respuesta JSON en un diccionario de Python
        data = response.json()

        results = data["data"]["home_search"]["results"]
        results = pd.DataFrame(results)
        # Define el nombre del archivo donde se guardará la respuesta JSON
        
    else:
        results=f"the request was failed. Error code: {response.status_code}"

    return results

```

```python
url2='https://es.wikipedia.org/wiki/Anexo:Ciudades_de_Estados_Unidos_por_poblaci%C3%B3n'
page=pd.read_html(url2)
State=page[1]
```

```python
State['Ranking.1']=(State['Ranking.1'].astype(str).str.split('-')).str[0]
```

```python
State=State.groupby('Ranking.1')['Ciudad'].apply(list).reset_index()
State.head()
```

```python
demand=['AL','AR','AZ','CA','CO','CT','FL','GA','HI','IA','ID','IL','IN','KS','KY','LA','MA','MD','MI','MN',
          'MO','NC','NE','NM','NV','NY','OH','OK','OR','PA','RI','SC','TN','TX','UT','VA','WA','WI']
```

```python
dfs=[]
for index,row in State.iterrows():
    
    State_code=row['Ranking.1']

    for city in row['Ciudad']:

        df=get_for_sale(State_code,city)
        

        dfs.append(df)


        
```

```python
df_final=pd.concat(dfs,ignore_index=True)
```

```python
from pulp import LpMaximize, LpProblem, LpVariable

# Grupo de números enlazados
grupo1 = [1, 2, 3, 4, 5]
grupo2 = [10, 20, 30, 40, 50]

# Crear un problema de PL para maximizar la suma en el segundo grupo
problem = LpProblem("Maximizar_Suma", LpMaximize)

# Variables de decisión binarias: 1 si se selecciona el número del primer grupo, 0 en caso contrario
seleccionados = LpVariable.dicts("Seleccionados", grupo1, cat="Binary")

# Función objetivo: Maximizar la suma en el segundo grupo
problem += sum(seleccionados[i] * grupo2[i] for i in range(len(grupo1))), "Suma_en_Grupo2"

# Restricción: Seleccionar máximo un número del primer grupo
problem += sum(seleccionados[i] for i in grupo1) <= 1, "Seleccion_Max_1_Numero"

# Resolver el problema
problem.solve()

# Obtener los números seleccionados
numeros_seleccionados = [i for i in grupo1 if seleccionados[i].varValue == 1]

# Imprimir resultados
print("Estado:", problem.status)
print("Números del primer grupo seleccionados:", numeros_seleccionados)
print("Suma máxima en el segundo grupo:", problem.objective.value())
```
